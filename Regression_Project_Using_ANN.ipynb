````python
import pandas as pd
import numpy as np
````

````python
# Boston Housing Price Prediction using Neural Networks (Keras)
# 1. Load the dataset


from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential,load_model
from tensorflow.keras.layers import Dense, Dropout
````

````python
from sklearn.datasets import fetch_california_housing
````

````python
df = fetch_california_housing()
````

````python
df #--Json format data
````

````python
type(df)
````

````python
X = pd.DataFrame(df.data,columns=df.feature_names)
````

````python
X
````

````python
y = pd.Series(df.target, name='MedHouseVal')
````

````python
y
````

````python
X.isna().sum()
````

````python
y.isna().sum()
````

````python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=24)
````

````python
X_train.shape
````

````python
y_train.shape
````

````python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
````

````python
# 3. Build the Neural Network Model
model = Sequential([
    Dense(64, activation = 'relu', input_shape = (X_train.shape[1],)),
    Dense(32, activation = 'relu'),
    Dense(1) # Regression output
])

model.compile(optimizer='adam', loss='mse',metrics = ['mae'])
````

````python
#4. Train the model
history = model.fit(X_train_scaled,y_train,validation_split= 0.2,epochs = 10, batch_size = 32)
````

### Neural Network Architecture Diagram

Below is a simple diagram representing the architecture defined in your code:

```
Input Layer (features)
      |
      v
+-------------------+
|   Dense Layer     |   (64 neurons, ReLU)
+-------------------+
      |
      v
+-------------------+
|   Dense Layer     |   (32 neurons, ReLU)
+-------------------+
      |
      v
+-------------------+
|   Output Layer    |   (1 neuron, regression output)
+-------------------+
```

- **Input Layer:** Number of neurons = number of features in your data (e.g., 8 for California housing).
- **First Hidden Layer:** 64 neurons, ReLU activation.
- **Second Hidden Layer:** 32 neurons, ReLU activation.
- **Output Layer:** 1 neuron, outputs the predicted value (house price).

This is a feedforward neural network (fully connected, sequential).

````python
# 5. Save the Model
model.save('deep_learning_model_reg.keras')
````

````python
# 6. Load the Model and Predict on New Data
loaded_model = load_model('deep_learning_model_reg.keras')
predictions = loaded_model.predict(X_test_scaled)
````

````python
# 7. Evaluate the Model
from sklearn.metrics import mean_squared_error, mean_absolute_error
mse = mean_squared_error(y_test, predictions)
mae = mean_absolute_error(y_test, predictions)
print(f"Test MSE: {mse:.2f}")
print(f"Test MAE: {mae:.2f}")
````
