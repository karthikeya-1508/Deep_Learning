padding(sequences to make them of same length): 

Means we need to 
make all the sequences of same length, 
because in deep learning model we need to have fixed input size.

review 1: [1, 2, 3, 4]  -- input shape is 4
        review 2: [4, 3, 6, 7,    input shape is 8
        
We can't give input shape of 4 and 8 to our deep learning model.

We need to give one fixed input shape to our model.

1. either we can truncate the longer sequences to make them of same length as shorter sequences.

    review 1: [1, 2, 3, 4]  -- input shape is 4
    review 2: [4, 3, 6, 7]   input shape is 4

2. or we can pad the shorter sequences with zeros to make them of same length as longer sequences.

    review 1: [0, 0, 0, 1, 2, 3, 4]  -- input shape is 8
    review 2: [4, 3, 6, 7, 8, 9, 10] input shape is 8

3.post padding: adding zeros at the end of the sequence

    review 1: [1, 2, 3, 4, 0, 0, 0]  -- input shape is 8
    review 2: [4, 3, 6, 7, 8, 9, 10] input shape is 8

```python
# 1. Data Collection
texts = [
    "I love this movie!",
    "This film was terrible.",
    "Amazing acting and story.",
    "I hated the ending.", 
    "Best movie of the year!",
    "Not worth watching."
]
labels = [1, 0, 1, 0, 1, 0]  # 1: positive, 0: negative
```

```python
texts
```

```python
import re

def clean_text(text):
  text = text.lower()
  text = re.sub(r'[^a-zA-z0-9\s]','',text)
  return text
text_cleaned = [clean_text(text) for text in texts]
```

```python
text_cleaned
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import Dense, Embedding, Flatten
```

```python
'''
Keras Tokenizer

if we have 1000 unique words in our corpus, Keras tokenizer will assign a unique integer to each word from 1 to 1000.
How it will assign the integers to each word?
it will assign the integers based on the frequency of the words in the corpus.

Movie -- 10 times -- 1
This -- 8 times -- 2
is -- 7 times -- 3
next freq word --4

last freq word -- 1000
'''
```

```python
# i don't wantto consider top words, I need to consider all teh words from the corpus
tokenizer = Tokenizer(
    num_words=None,           # Max number of words to keep, based on word frequency
    filters='!"#$%&()*+,-./:;<=>?@[\\]^_`{|}~\t\n',  # Characters to filter out
    lower=True,               # Convert texts to lowercase
    split=' ',                # Separator for splitting words
    char_level=False,         # If True, every character will be treated as a token
    oov_token=None,           # Token for out-of-vocabulary words
    document_count=0          # Number of documents (auto set during fitting)
)
tokenizer.fit_on_texts(text_cleaned)
sequences = tokenizer.texts_to_sequences(text_cleaned)
```

```python
sequences
```

```python
from keras.preprocessing.sequence import pad_sequences
max_len = 6
X = pad_sequences(sequences,maxlen = max_len, padding = 'post')
```

```python
X
```

```python
y = labels
```

```python
y
```

```python
#Update the code without embedding and flatten. just want to see the accuracy without that
# this is ANN architecture
model = Sequential([
    Dense(8, activation='relu', input_shape=(max_len,)),
    Dense(1, activation='sigmoid')
])
#Above archotecture with one input layer and one hidden layer and one output layer
```

```python
'''
# this is DNN architecture
#model = Sequential([
#    Dense(8, activation='relu', input_shape=(maxlen,)),
#    Dense(8, activation='relu'),
#    Dense(1, activation='sigmoid')
'''
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# 6. Model Training
import numpy as np
history = model.fit(np.array(X), np.array(y), epochs=10, batch_size=2, verbose=1)
```

```python
# 7. Model Evaluation
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

y_pred_prob = model.predict(np.array(X))
y_pred = (y_pred_prob > 0.5).astype(int).flatten()
```

```python
print('Accuracy:', accuracy_score(y, y_pred))
print('Precision:', precision_score(y, y_pred))
print('Recall:', recall_score(y, y_pred))
print('F1-score:', f1_score(y, y_pred))
print('Confusion Matrix:\n', confusion_matrix(y, y_pred))
```
